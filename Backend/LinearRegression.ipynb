```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv('C:/Users/mdasi/Desktop/Travel_Ease/Travel-Trails/Backend/tourism_total_cost_dataset1.csv')
```

```python
df.head()
```

```python
dummies=pd.get_dummies(df['State Name'])
```

```python
dummies
```

```python
merged=pd.concat([df,dummies],axis=1)
```

```python
merged
```

```python
df = merged.drop(['State Name'], axis=1)
```

```python
df.head()
```

```python

features = ['Hotel Price', 'Self Allowance', 'Travel Price', 'Food Cost', 'Goa', 'Kerala', 'Rajasthan', 'Sikkim','West Bengal']

x = df[features]

y = df['Total Cost']

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_pred)
```
